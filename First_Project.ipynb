```python
import json
import openpyxl
from openpyxl.styles import PatternFill , Font

with open('tech_data.json', 'r') as f:
    data = json.load(f)

low_stock = []
for item in data['inventory']:
    if data['inventory'][item] <10:
        low_stock.append(item)
print(low_stock)


print("Items running low on stock:", low_stock)

max_revenue = 0
top_product = ""
for curr_rev in data['sales_revenue']:
    if data['sales_revenue'][curr_rev]>max_revenue:
        max_revenue = data['sales_revenue'][curr_rev]
        top_product = curr_rev
print(max_revenue)

print("Top selling product is:", top_product)
wb =openpyxl.Workbook()
ws =wb.active
ws.title ="Overview"
ws['A1']="Company Name"
ws['B1']=data['company_info']['name']
ws['A3']="Top Product"
ws['B3'] = top_product
ws_inventory =wb.create_sheet("Inventory Alerts")
ws_inventory['A1'] = "Low Stock Items"
row_num = 2
for item in low_stock:
    ws_inventory.cell(row = row_num,column = 1,value =item)
    row_num =row_num + 1 
red_fill = PatternFill(start_color ='FF0000', fill_type = 'solid')
white_bold_font =Font(color='FFFFFF',bold=True)
ws_inventory['A1'].fill =red_fill
ws_inventory['A1'].font = white_bold_font   
wb.save ("Tech_Dashboard.xlsx")
print("Dashboard updated with Inventory Alerts!")
```
