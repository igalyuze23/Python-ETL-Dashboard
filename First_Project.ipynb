```python
import json
import openpyxl
from openpyxl.styles import PatternFill, Font

# --- שלב 1: טעינת הנתונים ---
# פתיחת קובץ ה-JSON שבו שמורים נתוני המלאי והמכירות
with open('tech_data.json', 'r') as f:
    data = json.load(f)

# --- שלב 2: ניתוח המלאי (Inventory Analysis) ---
low_stock = []
# מעבר על כל הפריטים במילון המלאי
for item in data['inventory']:
    # בדיקה: אם כמות המלאי קטנה מ-10, נוסיף את הפריט לרשימת התראות
    if data['inventory'][item] < 10:
        low_stock.append(item)

print("Items running low on stock:", low_stock)

# --- שלב 3: זיהוי המוצר הכי רווחי (Top Seller) ---
max_revenue = 0
top_product = ""
# מעבר על נתוני ההכנסות כדי למצוא את הערך הגבוה ביותר
for curr_rev in data['sales_revenue']:
    if data['sales_revenue'][curr_rev] > max_revenue:
        max_revenue = data['sales_revenue'][curr_rev]
        top_product = curr_rev

print("Top selling product is:", top_product)

# --- שלב 4: יצירת דוח אקסל (Excel Dashboard) ---
wb = openpyxl.Workbook() # יצירת קובץ אקסל חדש בזיכרון
ws = wb.active # גישה לגיליון הראשון (ברירת המחדל)
ws.title = "Overview" # שינוי שם הגיליון ל"סקירה כללית"

# הכנסת נתוני החברה והמוצר המוביל לגיליון הראשון
ws['A1'] = "Company Name"
ws['B1'] = data['company_info']['name']
ws['A3'] = "Top Product"
ws['B3'] = top_product

# --- שלב 5: יצירת גיליון התראות מלאי ---
ws_inventory = wb.create_sheet("Inventory Alerts") # יצירת גיליון חדש נוסף
ws_inventory['A1'] = "Low Stock Items"

# לולאה שכותבת את כל הפריטים החסרים שורה אחרי שורה
row_num = 2
for item in low_stock:
    ws_inventory.cell(row=row_num, column=1, value=item)
    row_num = row_num + 1 

# --- שלב 6: עיצוב ויזואלי (Styling) ---
# הגדרת צבע מילוי אדום ופונט לבן מודגש לכותרת ההתראות
red_fill = PatternFill(start_color='FF0000', fill_type='solid')
white_bold_font = Font(color='FFFFFF', bold=True)

# החלת העיצוב על תא הכותרת בגיליון המלאי
ws_inventory['A1'].fill = red_fill
ws_inventory['A1'].font = white_bold_font   

# --- שלב 7: שמירת הקובץ הסופי ---
wb.save("Tech_Dashboard.xlsx")
print("Dashboard updated with Inventory Alerts!")
```
